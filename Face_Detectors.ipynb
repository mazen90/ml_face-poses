```python
from math import sin , cos
def draw_axis(img, pitch,yaw,roll, tdx=None, tdy=None, size = 100):

    yaw = -yaw
    if tdx != None and tdy != None:
        tdx = tdx
        tdy = tdy
    else:
        height, width = img.shape[:2]
        tdx = width / 2
        tdy = height / 2

    # X-Axis pointing to right. drawn in red
    x1 = size * (cos(yaw) * cos(roll)) + tdx
    y1 = size * (cos(pitch) * sin(roll) + cos(roll) * sin(pitch) * sin(yaw)) + tdy

    # Y-Axis | drawn in green
    #        v
    x2 = size * (-cos(yaw) * sin(roll)) + tdx
    y2 = size * (cos(pitch) * cos(roll) - sin(pitch) * sin(yaw) * sin(roll)) + tdy

    # Z-Axis (out of the screen) drawn in blue
    x3 = size * (sin(yaw)) + tdx
    y3 = size * (-cos(yaw) * sin(pitch)) + tdy

    cv2.line(img, (int(tdx), int(tdy)), (int(x1),int(y1)),(0,0,255),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x2),int(y2)),(0,255,0),3)
    cv2.line(img, (int(tdx), int(tdy)), (int(x3),int(y3)),(255,0,0),2)

    return img
```

```python
import os
count = 0
ocount = 0
mesh = mp.solutions.face_mesh.FaceMesh(static_image_mode=True)
samples = []
imgs = os.listdir(images_path)
for i in imgs:
  ocount+=1
  print(ocount)
  print(images_path+'/'+i)
  im = cv2.imread(images_path+'/'+i)
  print(im.shape)
  imb = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
  ft = mesh.process(imb)
  lx=[]
  ly=[]
  # print((ft.multi_face_landmarks)[0].landmark[0])
  # break
  
  for j in range(468):
    if(ft.multi_face_landmarks) == None:
      break
    lx.append((ft.multi_face_landmarks)[0].landmark[j].x)
    ly.append((ft.multi_face_landmarks)[0].landmark[j].y)
    # print(j.landmark[0],j.landmark[0])
    count+=1
    # print(count)
  samples.append(np.hstack((np.array(lx),np.array(ly))))  
    
```
